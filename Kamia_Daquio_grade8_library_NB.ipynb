```python
import requests
import json

url = "https://raw.githubusercontent.com/YunaVDaquio/Kamia-Daquio-grade8-library-Repo/refs/heads/main/books.json"

response = requests.get(url)
books_data = response.json()

print(books_data)  # Verify it loaded
```

```python
import firebase_admin
from firebase_admin import credentials, db

# Load credentials
cred = credentials.Certificate("firebase_key.json")

firebase_admin.initialize_app(cred, {
    'databaseURL': 'https://kamia-daquio-grade8-libr-9a6d8-default-rtdb.asia-southeast1.firebasedatabase.app/'
})

print("Connected to Firebase!")
```

```python
ref = db.reference("books")

ref.set(books_data)

print("Books uploaded successfully!")
```

```python
ref = db.reference("/")
ref.set(books_data)
```

```python
ref = db.reference("books")
data = ref.get()

print(data)
```

```python
for book in data:
    print(book)
```

```python
for book in data:
    print(book["title"])
```

```python
for book in data:
    print("Title:", book["title"])
    print("Available:", book["details"]["available"])
    print("-----------")
```

```python
total_books = len(data)
print("Total number of books:", total_books)
```
